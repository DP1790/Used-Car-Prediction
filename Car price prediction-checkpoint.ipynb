##### We have data noisy in column brand, model & year. Lets fix this

```python
print(df['Brand'].nunique())
df['Brand'].value_counts()
```

```python
df['Brand']=df['Brand'].apply(lambda x:x.split(' ')[0])
```

```python
# Cleaning Brand Column
import re
brand=[]
for i in df['Brand']:
    x=re.sub(r'[0-9]','',i)
    brand.append(x)
len(brand)
```

```python
df['Brand']=brand #after cleaning new values of Brand
df
```

```python
#cleaning noises of column model
Model=[]
for i in df['model']:
    i=i.split()
    if len(i)>=3:
        Model.append(i[1])
    else:
        Model.append(''.join(i))
len(Model)
```

```python
df['Model']=Model
```

```python
df['Model']=df['Model'].replace('R','Wagon R') #replace R with Wagon R as during cleaning Wagon is deleted
df
```

```python
#Cleaning noises in column Km
Km=[]
for i in df['km driven']:
    i=re.sub(r'[A-Za-z]+','',i)
    i=re.sub('\s+','',i)
    Km.append(i)
Km
```

```python
df['Km Driven']=Km
df
```

```python
df['Km Driven']=df['Km Driven'].apply(lambda x:int(x.replace(',','')))#convert Km driven column to numeric
df
```

```python
df=df.drop(['model','km driven'],axis=1) # dropping unwanted columns as we have created separate columns
df
```

```python
#As we can see 1st,First owner, 1st Owner and First owner represent single value & similarly others as well. we have to make all to same unique value.
df['No of owners']=df['No of owners'].replace('1st','First Owner')
df['No of owners']=df['No of owners'].replace('First','First Owner')
df['No of owners']=df['No of owners'].replace('1st Owner','First Owner')
df['No of owners']=df['No of owners'].replace('2nd','Second Owner')
df['No of owners']=df['No of owners'].replace('Second','Second Owner')
df['No of owners']=df['No of owners'].replace('2nd Owner','Second Owner')
df['No of owners']=df['No of owners'].replace('3rd','Third Owner')
df['No of owners']=df['No of owners'].replace('3rd Owner','Third Owner')
df['No of owners']=df['No of owners'].replace('4th','Fourth Owner')
df['No of owners']=df['No of owners'].replace('Fourth','Fourth Owner')
```

```python
df['price'].unique()
```

```python
Price=[]
for i in df['price']:
    i=i.split()
    if len(i)==2:
        if i[-1]=='Lakh':
            Price.append(float(i[0])*100000)#where price value is lakh we multiply with 100000
        elif i[-1]=='Cr':
            Price.append(float(i[0])*10000000)#where price value is cr we multiply with 10000000
        else:
            pass
    else:
        Price.append(''.join(i))
len(Price)
```

```python
df['Price']=Price
df
```

```python
## Divide data set into features and label
y = df['Price']
x = df.drop(columns=['Price'], axis=1)
```

##### Finding best Random_ state

```python
from sklearn.ensemble import RandomForestRegressor
maxr2score = 0
maxRS = 0

for i in range (1, 150):
    x_train, x_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.2, random_state = i)
    RF = RandomForestRegressor()
    RF.fit(x_train, y_train)
    predrf = RF.predict(x_test)
    r2 = r2_score(y_test, predrf)
    if r2>maxr2score:
        maxr2score = r2
        maxRS = i
print ('Max R2 score obtained is', maxr2score, 'on Random State', maxRS)
```

##### Creating train test split

```python
# Split the data into train and test. Model will be built on training data and tested on test data.
x_train,x_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.2, random_state=maxRS)
y_train.head()
```

##### Hyper parameter tuning

```python
params = {'n_estimators': range(50,200,10),
          'criterion' : ['mse', 'mae'],
               'max_features': ['auto', 'sqrt','log2'],
               'max_depth': range(5,30,6),
               'min_samples_split': [2, 5, 10, 15, 100],
               'min_samples_leaf': [1, 2, 5, 10]}
```

```python
from sklearn.model_selection import RandomizedSearchCV
rfr = RandomizedSearchCV(RandomForestRegressor(),params,n_iter = 10,cv=5, n_jobs= -1,random_state=maxRS)
rfr.fit(x_train,y_train)
```
