# Web Scrapping: Used Car Price

```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import ElementClickInterceptedException
from bs4 import BeautifulSoup
import requests
import time
```

```python
driver = webdriver.Chrome('chromedriver.exe')
time.sleep(2)
url="https://www.olx.in/"
driver.get(url)
```

```python
driver.find_element_by_xpath("//a[@class='_2fitb']").click()
```

```python
for i in range(50):
    try:
        driver.execute_script("window.scrollBy(0,40000)")
        time.sleep(1)
        driver.find_element_by_xpath("//button[@class='rui-39-wj rui-3evoE rui-1JPTg']").click()
        time.sleep(1)
    except NoSuchElementException as e:
        print(e)
        break
```

```python
urls=[]
for i in driver.find_elements_by_xpath("//li[@class='EIR5N']/a"):
    urls.append(i.get_attribute("href"))
```

```python
len(urls)
```

```python
#Lists
brand=[]
model=[]
variant=[]
year=[]
fuel=[]
transmission=[]
km=[]
owners=[]
location=[]
price=[]
```

```python
for i in urls:
    driver.get(i)
    time.sleep(1)
    try:
        brand.append(driver.find_element_by_xpath("//div[@class='_35xN1']").text)
    except NoSuchElementException:
        brand.append('-')      
    try:
        model.append(driver.find_element_by_xpath("//div[@class='_35xN1']").text)
    except NoSuchElementException:
        model.append('-')
    try:
        year.append(driver.find_element_by_xpath("//div[@class='_35xN1']").text)
    except NoSuchElementException:
        year.append('-')
    try:
        variant.append(driver.find_element_by_xpath("//div[@class='_3tLee']").text)
    except NoSuchElementException:
        variant.append('-')
    try:
        fuel.append(driver.find_element_by_xpath("/html/body/div/div/main/div/div/div/div[5]/div[1]/div[1]/div[1]/div/div[4]/div[1]/div").text)
    except NoSuchElementException:
        fuel.append('-')
    try:
        transmission.append(driver.find_element_by_xpath("/html/body/div[1]/div/main/div/div/div/div[5]/div[1]/div[1]/div[1]/div/div[4]/div[3]/div").text)
    except NoSuchElementException:
        transmission.append('-')
    try:
        km.append(driver.find_element_by_xpath("/html/body/div/div/main/div/div/div/div[5]/div[1]/div[1]/div[1]/div/div[4]/div[2]/div").text)
    except NoSuchElementException:
        km.append('-')
    try:
        owners.append(driver.find_element_by_xpath("/html/body/div/div/main/div/div/div/div[5]/div[1]/div[2]/div/div[2]/div/div[1]/div[2]/div[2]").text)
    except NoSuchElementException:
        owners.append('-')
    try:
        location.append(driver.find_element_by_xpath("/html/body/div/div/main/div/div/div/div[5]/div[1]/div[2]/div/div[2]/div/div[2]/div[2]/div[2]").text)
    except NoSuchElementException:
        location.append('-')
    try:
        price.append(driver.find_element_by_xpath("//div[@class='_3FkyT']").text)
    except NoSuchElementException:
        price.append('-')
```

```python
df=pd.DataFrame({})
df['Brand']=brand
df['model']=model
df['variant']=variant
df['year']=year
df['fuel']=fuel
df['transmission']=transmission
df['km driven']=km
df['No of owners']=owners
df['location']=location
df['price']=price
df
```

```python
df.to_csv("olx.csv")
```
